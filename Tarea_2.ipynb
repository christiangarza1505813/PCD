```python
import json
import gzip
import pandas as pd

file = '/content/Video_Games.jsonl.gz'

data = []  # Lista donde guardaremos cada objeto JSON

try:
    # Intentar abrir el archivo como JSON Lines "normal"
    with open(file, 'r', encoding='utf-8') as fp:
        for line in fp:
            data.append(json.loads(line.strip()))
except UnicodeDecodeError:
    # Si ocurre un problema de codificación, probamos abrirlo como gzip
    print("File appears to be gzip compressed. Attempting to decompress...")
    with gzip.open(file, 'rt', encoding='utf-8') as fp:
        for line in fp:
            data.append(json.loads(line.strip()))

# Convertir la lista de diccionarios en un DataFrame
df = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame para corroborar
df.head(100)
```

#Limpieza y tokenización

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize



```

```python
def tokenizar(text):
    text = text.replace('<br /><br />', ' ')
    text = text.replace('br', ' ')
    text = text.replace('game', ' ')
    texto_limpio = re.sub(r'[^\w\s]', '', text.lower())
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado




df_filtered['token'] = df_filtered['text'].apply(lambda x : tokenizar(x))
df_filtered['token2'] = df_filtered['text'].apply(lambda x : tokenizar(x))
```

```python
df_filtered = df_filtered[["rating", "text", "token", "token2"]]
df_filtered.head()
```

```python
df_filtered_ex = df_filtered.explode(column='token')
df_filtered_ex.head()
```

```python
lexicon = pd.read_table(
            'https://raw.githubusercontent.com/fnielsen/afinn/master/afinn/data/AFINN-en-165.txt',
            names = ['termino', 'sentimiento']
          )
# Transformar los valores de sentimiento de -5 a 5 a una escala de 1 a 5
lexicon['sentimiento_transformado'] = 1 + ((lexicon['sentimiento'] + 5) / 2.5)

# Verificar el resultado
print(lexicon.head())
lexicon.head()
```

```python
lexicon["sentimiento_transformado"].unique()
```

```python
df_filtered_ex['token2'] = df_filtered_ex['token2'].apply(lambda x: tuple(x) if isinstance(x, list) else x)
df_merge = (
    pd.merge(df_filtered_ex, lexicon, how='left', left_on='token', right_on='termino')
    .groupby(['rating', 'text','token2'])
    .sentimiento_transformado.mean()
    .reset_index()
    .sort_values(by='sentimiento_transformado', ascending=False)
    .dropna()
)
```

#Análisis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Calcular y mostrar correlación
correlation = df_merge[['rating', 'sentimiento_transformado']].corr()
print("Correlación entre rating y sentimiento:")
print(correlation)

# Diagrama de dispersión
sns.scatterplot(data=df_merge, x='rating', y='sentimiento_transformado')
plt.title('Relación entre Rating y Sentimiento')
plt.show()
```

```python
# Histograma de los sentimientos transformados
df_merge['sentimiento_transformado'].plot(kind='hist', bins=10, alpha=0.7, color='lightblue', label='Sentimientos')
plt.title('Distribución de Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Histograma de las calificaciones originales (rating)
df_merge['rating'].plot(kind='hist', bins=10, alpha=0.7, color='green', label='Calificaciones Originales')
plt.title('Distribución de Calificaciones Originales')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
avg_sentiment = df_merge.groupby('rating')['sentimiento_transformado'].mean().reset_index()

# Gráfico de barras
sns.barplot(data=avg_sentiment, x='rating', y='sentimiento_transformado')
plt.title('Promedio de Sentimiento por Calificación')
plt.xlabel('Rating')
plt.ylabel('Promedio de Sentimiento')
plt.show()
```

```python
# Filtrar discrepancias grandes
outliers = df_merge[(df_merge['rating'] >= 4) & (df_merge['sentimiento_transformado'] < 2)]
print("Reseñas con alta calificación pero bajo sentimiento:")
print(outliers.head(1))

# Visualizar con boxplot
sns.boxplot(data=df_merge, x='rating', y='sentimiento_transformado')
plt.title('Distribución de Sentimiento por Calificación')
plt.show()
```
